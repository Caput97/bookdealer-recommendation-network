```python
#function to obtain number of neighbors of a specific node
def n_of_neighbors(nodeID):
    countNeig = 0
    for neighbor in list(g.neighbors(nodeID)):
        countNeig += 1
    return countNeig
```

##### Most important nodes

```python
#try to retrieve the most important nodes through the intersection of the nodes found
# by using the different centrality measures above


ImpDegree = []
ImpCloseness = []
ImpBetweenness = []
ImpCloseness = []
ImpHarmonic = []
ImpPR = []
ImpEigenV = []

#function that takes the first 100 nodes in each rank and append them in a specific list.
#So we will have 6 lists w/ the 100 most important nodes for each centr. measure
def get_100nodes(rank, list):
    for el in rank[0:100]:
        list.append(el[0])

get_100nodes(ranksDeg, ImpDegree)
get_100nodes(ranksClos, ImpCloseness)
get_100nodes(ranksBet, ImpBetweenness)
get_100nodes(ranksHarm, ImpHarmonic)
get_100nodes(ranksPr, ImpPR)
get_100nodes(ranksEig, ImpEigenV)


#intersection between the 6 list in order to create a common list of the most important nodes of our network
MostImportantNodes = list(set(ImpDegree) & set(ImpCloseness) & set(ImpBetweenness) & set(ImpHarmonic) & set(ImpPR)  & set(ImpEigenV))

#first 10 most important nodes
MostImportantNodes[0:10]
```

```python
MINodes = pd.DataFrame(columns=['NodeID', 'BookName', 'Genre', 'Publisher', 'Year', 'Recommended by', 'Node Degree'])

for node in MostImportantNodes[0:10]:
    new_row = {'NodeID':node, 'BookName':name[node], 'Genre':genre[node], 'Publisher':publisher[node], 'Year':year[node], 'Recommended by':recommendations[node], 'Node Degree':g.degree(node)}
    MINodes = MINodes.append(new_row, ignore_index=True)

# Sort by Descending
MINodes = MINodes.sort_values('Node Degree', ascending=False)

#DF w/ the 10 most important nodes sorted by node degree value
MINodes
```
